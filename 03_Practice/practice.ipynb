### Task 

Write a function that:

Takes a string s

Loops through each character

Checks if the character is a vowel

Counts vowels

Prints the final count

Assume the string is lowercase.

```python
def vowel_count(word):
    count = 0
    vowels = ['a','e','i','o','u']
    for char in word.lower():
        if char in vowels:
            count += 1
    print(count)
        
```

```python
vowel_count("bcdf")
```

```python
vowel_count("PyThOn")
```

```python
def freq_count(word):
    freq={}
    for char in word:
        if char in freq:
            freq[char]+=1
        else:
            freq[char]=1
    return(freq)
```

### Task
Write a function that:

Builds the frequency dictionary (you already know this)

Loops through the string again

Prints the first character with frequency 1

If none found, prints something like "None"

```python
def non_rep_char(word):
    freq={}
    for char in word:
        if char in freq:
            freq[char]+=1
        else:
            freq[char]=1
    for char in word:
        if freq[char] == 1:
            return char
    return "None"
```

```python
non_rep_char("aabbcc")
```

### Task

Given a number n, loop from 1 to n and:

if the number is divisible by 5, skip it

if the number is divisible by 11, stop the loop

otherwise, add the number to a sum

```python
def multi_cond(n):
    total = 0
    for i in range(1,n+1):
        if i%5 == 0:
            continue
        elif i%11 == 0:
            break
        else:
            total += i
    print(total)
        
```
