```python
def dolistsum(number_list):
  total_sum = 0
  for num in number_list:
    total_sum += num
  print(total_sum)
```

```python
#Create function to check a number is prime or not

n = 9

if n > 1:
  for i in range(2,n):
    if n % i == 0:
      print(n , "is not a prime")
      break
  else:
    print(n , "is a prime")
else:
  print("given number is negative or 0 or 1")
```

```python
def is_prime(num):
  if num > 1:
    for i in range(2,num):
      if num % i == 0:
        print(num , "is not a prime")
        break
    else:
      print(num , "is a prime")
  else:
    print("given number is negative or 0 or 1")
```

```python
is_prime(53)
```

```python
#create a return type function which can check a number is even or odd

def even_or_odd(num):
  if num % 2 == 0:
    return True
  else:
    return False
```

```python
num = int(input("Enter a number to check even or odd"))

status = even_or_odd(num)

if status:
  print(num , "is Even")
else:
  print(num , "is Odd")
```

```python
#Create a function that can return list of all the vowels present in text

def identify_vowels(text):
  vowels = []
  for char in text:
    if char.lower() in ['a','e','i','o','u']:
      vowels.append(char)
  return vowels, len(vowels)
```

```python
text = "The Evening Is Great Today."

list_of_vowels , total_number_of_vowels = identify_vowels(text)

print(list_of_vowels)
print(total_number_of_vowels)
```
