**Inheritance**
  * Inheritance is fundamental fetaure of object-oriented programming that allows a class (called the child or subclass) to inherit attributes and methods from another class (called the parent or superclass). This promotes code reusability and allows for creation of hierarchical relationships between classes.

```python
class class_A:
  x = 7
  y = 5


class class_B(class_A):
  pass
```

```python
objA = class_A()
print(objA.x)
print(objA.y)
```

```python
objB = class_B()
print(objB.x)
print(objB.y)
```

```python
#Base class or Parent class
class students:
  def __init__(self,name,age):                     #Initialiser
    self.name = name
    self.age = age


#Derived class or Child class
class school_student(students):
  def Sstudent(self,standard):
    self.class_standard = standard
    print(f"{self.name} is of age {self.age} and studying in class {self.class_standard}")


#Derived class or Child class
class college_student(students):
  def Cstudent(self,year):
    self.year = year
    print(f"{self.name} is of age {self.age} and studying in {self.year} year")
```

```python
school_stu = school_student("Tarun",12)
school_stu.Sstudent('5th')
```

```python
college_stu = college_student("Ajay" , 20)
college_stu.Cstudent("3rd")
```

```python
print(school_stu.name , school_stu.age)
print(college_stu.name , college_stu.age)
```

```python
class Animal:
  def sound(self):
    print("Some Sound")


class Dog(Animal):
  def bark(self):
    print("Woof... Woof")
```

```python
#Base class 1
class employee:
  def emp_details(self,n,a):
    self.name = n
    self.age = a

#Base class 2
class department:
  def emp_dpt(self,dept):
    self.department = dept


#Derived class
class company_employees(employee,department):
  def details(self):
    detail = {"Name" : self.name , "Age" : self.age , "Department" : self.department}
    return detail
```

```python
#Base class 1
class employee:
  def __init__(self,n,a):
    self.name = n
    self.age = a

#------------------------------------------------------------------------------------------------------------------

#Base class 2
class department:
  def __init__(self,dept):
    self.department = dept

#------------------------------------------------------------------------------------------------------------------

#Derived class
class company_employees(employee,department):
  def __init__(self,n,a,d):
    employee.__init__(self,n,a)
    department.__init__(self,d)

  def details(self):
    detail = {"Name" : self.name , "Age" : self.age , "Department" : self.department}
    return detail
```

```python
obj3 = company_employees("Arun",25,"HR")
obj3.details()
```

```python
obj4 = company_employees("John",28,"Tester")
obj4.details()
```

****
* Multilevel Inheritance
****

```python
#Base class1
class Grandfather:
  def method_one(self):
    print("This is from 1st base class (GrandFather)")


#Second Base class inherting base class 1
class Father(Grandfather):
  def method_two(self):
    print("This is from middle class (Father)")

#Child class
class Son(Father):
  def method_three(self):
    print("This is from last class (Son)")
```

```python
obj = Son()

obj.method_three()
obj.method_two()
obj.method_one()
```

****
* Hierarchical Inheritance
****

```python
#Base class
class Baseclass:
  def method_one(self):
    print("This is from base class")


#child class 1
class childclass1(Baseclass):
  def method_two(self):
    print("This is from child class1")

#Child class2
class childclass2(Baseclass):
  def method_three(self):
    print("This is from child class2")
```

```python

obj1 = childclass2()

obj1.method_three()
obj1.method_one()
```

```python
obj2 = childclass1()

obj2.method_two()
obj2.method_one()
```

****
* Hybrid Inheritance
****

```python
class A:
  def method_a(self):
    print("Method A")


class B(A):
  def method_b(self):
    print("Method B")


class C(A):
  def method_c(self):
    print("Method C")


class D(B,C):
  def method_d(self):
    print("Method D")
```

```python
obj1 = D()

obj1.method_d()
obj1.method_b()
obj1.method_c()
obj1.method_a()
```

```python
obj2 = C()

obj2.method_c()
obj2.method_a()
```
