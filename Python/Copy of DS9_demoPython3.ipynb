```python
#Grading
#fetch marks
#90 or above => Grade A+
#80 to 90 => Grade A
#70 to 80 => Grade B+
#60 to 70 => Grade B
#50 to 60 => Grade C+
#40 to 50 => Grade C
#less then 40 fail

marks = float(input("Enter Your Marks"))

if marks >= 90:
  print("Grade A+")
elif marks >= 80:
  print("Grade A")
elif marks >= 70:
  print("Grade B+")
elif marks >= 60:
  print("Grade B")
elif marks >= 50:
  print("Grade C+")
elif marks >= 40:
  print("Grade C")
else:
  print("Fail")
```

```python
numbers = [6,3,5,9,2,7,1,4]

#seperate even and odd numbers from given list of numbers

evens = []
odds = []

for num in numbers:
  if(num % 2 == 1):
    print("Odd Number")
    odds.append(num)
  else:
    print("Even Number")
    evens.append(num)
```

```python
#WAP for login activity where we can give 3 attempts for login incase entered credentials are wrong

for i in range(0,3):
  userid = input("Enter User Id")
  userpass = input("Enter User Password")

  if (userid == 'admin' and userpass == 'password'):
    print("Login Successfull")
    break
  else:
    print("Invalid login")
```

```python
for i in range(3,0,-1):
  userid = input("Enter User Id")
  userpass = input("Enter User Password")

  if (userid == 'admin' and userpass == 'password'):
    print("Login Successfull")
    break
  else:
    print("Invalid login")
    print("Number of Attempts left =" , i-1)
```
