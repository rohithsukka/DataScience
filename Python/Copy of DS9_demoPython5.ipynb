```python
#positional parameters or positional arguments

def show_details(name,age):
  details = {"Name" : name , "Age" : age}
  return details
```

```python
show_details("Arun" , 21)
```

```python
show_details(21 , "Arun")
```

```python
#Keyword arguments

show_details(age = 21 , name  = "Arun")
```

```python
#Default Parameter

def show_details(name , age = 'NA'):
  details = {"Name" : name , "Age" : age}
  return details

print(show_details("Arun"))
print(show_details("Arun",22))
```

```python
#Variable length parameter

def dosum(*args):
  total_sum = 0
  for val in args:
    total_sum += val
  return total_sum


print(dosum(4,3,5))
print(dosum(1,4,9,6))
print(dosum(4,3,5,2,9,7,4))
```

```python
#Variable length keyword parameters

def show_details(**kwargs):
  details = {}
  for key,arg in kwargs.items():
    details[key] = arg
  return details
```

```python
print(show_details(name="Pavan" , age = 21))
print(show_details(name="John" , age = 31 , email = "john@abc.com"))
print(show_details(name="Bipul"))
```

```python
#Create a function to check a number is even or not
def evenornot(num):
  return num % 2 == 0

print(evenornot(7))
print(evenornot(8))
```

```python
evenorodd = lambda num : num % 2 == 0
evenorodd(9)
```

```python
def evenornot(num):
  if num % 2 == 0:
    return "Even Number"
  else:
    return "Odd Number"

print(evenornot(7))
print(evenornot(8))
```

```python
evenorodd = lambda num : "Even Number" if num % 2 == 0 else "Odd Number"
evenorodd(22)
```

```python
def weight_measure(weight):
  if weight < 50:
    return "Less Weight"
  elif weight < 75:
    return "Optimum Weight"
  else:
    return "Heavy Weight"
```

```python
weight_measure(60)
```

```python
wgt_msr = lambda weight : "Less Weight" if weight < 50 else "Optimum Weight" if weight < 75 else "Heavy Weight"
wgt_msr(34)
```

```python
#Create a lambda function for login activity

Login = lambda username, password : "Login Successful" if username == "admin" and password == "admin" else "Login Failed"

username = input("Enter UserName")
userpassword = input("Enter User Password")

Login(username,userpassword)
```

**List Comprehension**

```python
numbers = []

for i in range(0,10):
  numbers.append(i)

print(numbers)
```

```python
nums = [i for i in range(0,10)]
print(nums)
```

```python
even_numbers = []

for i in range(0,20):
  if i % 2 == 0:
    even_numbers.append(i)

print(even_numbers)
```

```python
en = [i for i in range(0,20) if i % 2 == 0]
print(en)
```

```python
#create a list of square of numbers from 1 to 5 using list comprehension
squared_numbers = [i**2 for i in range(1,6)]
print(squared_numbers)
```

```python
fruits = ["apple" , "banana" , "kiwi"]

word_length = []

for fruit in fruits:
  word_length.append(len(fruit))
```

```python
print(word_length)
```

```python
[len(fruit) for fruit in fruits]
```

```python
[(fruit,len(fruit)) for fruit in fruits]
```

```python
celsius_temp = [0,10,20,30]

#temp_fah = (temp * 9/5) + 32

temp_f = [(temp * 9/5) + 32 for temp in celsius_temp]
print(temp_f)
```

```python
#Seperate vowels from given sentence using list comprehension

sentence = "Hello , How are you today"

filtered_volwels = [char for char in sentence if char.lower() in 'aeiou']
print(filtered_volwels)
```
