**File Handling in Python**
  * reading from and writing to files using python programming.
  * to open a file :- open("file_location" , "action-r/w/a/x")
      * **r** - read a file if it exist else generate error
      * **w** - write into a file if it exist or create if it not existing.
      * **a** - append new information into an existing file or create a new one if it is not existing.
      * **x** - create and write only if file is not existing

```python
#Write a file - if mentioned file doesn't exist=> create a new file and write into it

file = open("myfile.txt" , "w")
file.write("My text File using Python")
file.close()
```

```python
#Read a file

file = open("myfile.txt" , "r")
content = file.read()
print(content)
file.close()
```

```python
#Write a file - if mentioned file is existing => overwrite the old content within the file

file = open("myfile.txt" , "w")
file.write("Rewritting my text file using 'Python Scripting'")
file.close()
```

```python
#Read a file

file = open("myfile.txt" , "r")
content = file.read()
print(content)
file.close()
```

```python
#Append a text into existing file -

file = open("myfile.txt" , "a")
file.write("\nAdding some new content into the file")
file.close()
```

```python
#'a' can also create a new file if mentioned file is not existing

file = open("mynewfile.txt" , "a")
file.write("\nABCDEFGHIJKLMNOPQRSTUVWXYZ")
file.close()
```

```python
# 'x' - can create and write a new file, can't operate on existing file

file = open("mfile.txt" , "x")
file.write("Create and adding text using 'x'")
file.close()
```

```python
#Accessing ech line of a file one by one
file = open("myfile.txt" , "r")

while True:
  content = file.readline()
  if content == '':
    print("------Content Ends Here------")
    break
  print(content)

file.close()
```

```python
try:
  file = open("abcfile.txt" , "r")
  print(file.read())

except Exception as err:
  print(err)

finally:
  file.close()
```

```python
user_input_list = []
while True:
  user_input = input("Enter any text or number or bool value or 'exit' to take exit ")
  if user_input == 'exit':
    break
  try:
    if '.' in user_input:
      user_input_list.append(float(user_input))
    else:
      user_input_list.append(int(user_input))
  except:
    if user_input in ['True' , 'False']:
      user_input_list.append(eval(user_input))
    else:
      user_input_list.append(user_input)
```

```python
print(user_input_list)
```

* Raise custom exception

```python
def divide(x,y):
  if y == 0:
    raise ValueError("Cannot divide by zero")
  z = x / y
  return z
```

```python
try:
  result = divide(55,0)
except Exception as err:
  print(err)
```
