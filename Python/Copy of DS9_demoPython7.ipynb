```python
class myclass:
  x = 123
  y = 22
```

```python
obj = myclass()
print(obj.x)
print(obj.y)
```

```python
class mynewclass:
  x = 11                #class variables
  y = 22

  def do_sum(self):         #class method
    a = 6
    b = 3
    c = a/b
    print(c)
```

```python
obj = mynewclass()

print(obj.x)
print(obj.y)
```

```python
obj.do_sum()
```

```python
class Employee:
  company = "ABC"

  def emp_details(self):
    name = "Rahul"
    age = 22
    return {"Name" : name , "Age" : age}
```

```python
emp = Employee()

print(emp.company)
print(emp.emp_details())
```

```python
class Employee:
  company = "ABC"

  def emp_details(self):
    self.name = "Rahul"
    self.age = 22

  def show_emp_details(self):
    return {"Name" : self.name , "Age" : self.age}
```

```python
emp = Employee()                    #Object of class

print(emp.company)                  #class variable

emp.emp_details()                   #class method to define name and age variable
print(emp.show_emp_details())       #class method to show employee details
```
