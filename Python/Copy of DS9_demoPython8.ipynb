Object-Oriented Programming - Way of creating more structured & easy to debug program with two most important entity of programming -
  * attribute
  * behaviour
    * It can be achieved by creating class in python.

* attributes or properties - these are variables defined within a class.

```python
class myclass:
  name = "Raju"
  age = 22
```

```python
c1 = myclass()
```

```python
print(c1.name)
print(c1.age)
```

* behaviour or method - these are the functions defined within a class to perform certain activity

```python
class newclass:
  def docalc(self):
    x = 7
    y = 2
    z = x/y
    print(z)
```

```python
c2 = newclass()

c2.docalc()
```

```python
class Employees:
  company = 'ABC'

  def emp_details(self,n1,a1):
    self.name = n1
    self.age = a1

  def emp_email(self,e1):
    self.email = e1

  def show_details(self):
    return {"Name" : self.name , "Age" : self.age , "Email" : self.email}
```

```python
e1 = Employees()
e2 = Employees()
e3 = Employees()
```

```python
e3.emp_details("Arun",32)
e3.emp_email("arun@hello.com")
e3.show_details()
```

```python
print(e1.name, e1.age, e1.email)
print(e2.name, e2.age, e2.email)
print(e3.name, e3.age, e3.email)
```

```python
class Employees:
  company = 'ABC'
  count = 0              #class variable

  def emp_details(self,n1,a1):
    self.name = n1
    self.age = a1
    Employees.count += 1

  def emp_email(self,e1):
    self.email = e1

  def show_details(self):
    return {"Name" : self.name , "Age" : self.age , "Email" : self.email}
```

```python
e1 = Employees()
e2 = Employees()
e3 = Employees()
```

```python
e1.emp_details("Ajay" , 21)
e1.emp_email("ajay@hello.com")
e1.show_details()
print("Total Count =" , e1.count)
```

```python
e2.emp_details("Arun" , 27)
e2.emp_email("arun@hello.com")
e2.show_details()
print("Total Count =" , e2.count)
```

* create a Employee class with details name, age and email. Define a variable id and use autoincrement for id to define unique id for each employee.

```python
class Employees:
  company = 'ABC'
  id = 0              #class variable

  def emp_details(self,n1,a1):
    self.name = n1
    self.age = a1
    Employees.id += 1
    self.iid = Employees.id

  def emp_email(self,e1):
    self.email = e1

  def show_details(self):
    return {"Id" : self.iid, "Name" : self.name , "Age" : self.age , "Email" : self.email , "Company" : self.company}
```

```python
e1 = Employees()
e2 = Employees()
e3 = Employees()
```

```python
e1.emp_details("Ajay" , 21)
e1.emp_email("ajay@hello.com")

e2.emp_details("Arun" , 27)
e2.emp_email("arun@hello.com")

e3.emp_details("John" , 37)
e3.emp_email("john@hello.com")
```

```python
e1.show_details()
```
